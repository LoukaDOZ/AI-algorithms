```python
f = open("word.pl", "r")
words = []

for s in f.read().split('\n'):
    if len(s) >= 18:
        w = s[6] + s[9] + s[12] + s[15] + s[18]
        words.append(w)
    
print(words)
```

```python
count = {}

for w in words:
    for l in w:
        if l not in count:
            count[l] = 1
        else:
            count[l] += 1

count = dict(sorted(count.items(), key=lambda item: item[1], reverse = True))
print(count)
```

```python
sort = []
for w in words:
    c = 0
    for l in w:
        c += count[l]
    sort.append((w, c))
```

```python
def first(l, n):
    for i,v in enumerate(l):
        if v == n:
            return i
        
    return -1

def score_from_letter(l):
    for i,v in enumerate(count):
        if v == l:
            return len(count) - i
    return 0

def score(w):
    c = []
    for l in w:
        c.append('?')
    return score_rec(w, c)
    
def score_rec(w, c):
    #print(c)
    s = 0
    if '?' not in c:
        for i,v in enumerate(c):
            if v == 'g':
                s += 3 * score_from_letter(w[i])
            elif v == 'b':
                s += 2 * score_from_letter(w[i])
            else :
                s += score_from_letter(w[i])
        return s
    
    i = first(c, '?')
    
    if c.count('?') != 1 or c.count('o') != 1:
        g = c.copy()    
        g[i] = 'g'
        #print(g)
        s += score_rec(w, g)
    
    if i != len(c) - 1:
        o = c.copy()
        o[i] = 'o'
        #print(o)
        s += score_rec(w, o)
  
    b = c.copy()
    for j in range(0, len(w)):
        if w[j] == w[i]:
            b[j] = 'b'
    #print(b)
    s += score_rec(w, b)
        
    #print()
    #print()
    return s
 
print(score("abaca"))
print(score("tarie"))

sort = []
for w in words:
    sort.append((w, score(w)))

print(sort)
```

```python
def generate_cases(w, c=['?', '?', '?', '?', '?']):
    tab = []
    
    if '?' not in c:
        tab.append(c)
    else:
        i = first(c, '?')

        if c.count('?') != 1 or c.count('o') != 1:
            g = c.copy()    
            g[i] = 'g'
            #print(g)
            tab = tab + generate_cases(w, g)

        if i != len(c) - 1:
            o = c.copy()
            o[i] = 'o'
            #print(o)
            tab = tab + generate_cases(w, o)

        b = c.copy()
        for j in range(0, len(w)):
            if w[j] == w[i]:
                b[j] = 'b'
        #print(b)
        tab = tab + generate_cases(w, b)
        
    return tab

def allowed_and_forbidden(w, c):
    af = {
        "keep": [],
        "swap": {},
        "forbidden": []
    }
    
    for i,v in enumerate(c):
        if v == 'g':
            af["keep"].append((i, w[i]))
        elif v == 'o':
            if w[i] not in af["swap"]:
                af["swap"][w[i]] = {
                    "count": 1,
                    "not": [i]
                }
            else:
                af["swap"][w[i]]["count"] += 1
                af["swap"][w[i]]["not"].append(i)
        else:
            if w[i] not in af["swap"]:
                af["forbidden"].append(w[i])
            
    return af

def check(w, c):
    af = allowed_and_forbidden(wo, c)

    for i in af["keep"]:
        if w[i[0]] != i[1]:
            return False
    
    for i in af["swap"]:
        if w.count(i) != af["swap"][i]["count"]:
            return False
        
        for j in af["swap"][i]["not"]:
            if w[j] == i:
                return False
        
    for i in af["forbidden"]:
        if w.count(i) > 0:
            return False
            
    return True
    
sort = []
for i,w in enumerate(words):
    gc = generate_cases(w)

    total = []
    for c in gc:
        still = 0
        for w2 in words:
            if w != w2 and check(w2, c):
                still += 1
        total.append(still)
    sort.append((w, total))
    print("Done : ", i + 1, " / ", len(words))
```
