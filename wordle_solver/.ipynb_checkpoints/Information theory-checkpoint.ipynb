```python
f = open("word.pl", "r")
words = []

for s in f.read().split('\n'):
    if len(s) >= 18:
        w = s[6] + s[9] + s[12] + s[15] + s[18]
        words.append(w)
    
#print(words)
```

```python
def availables(p, s):
    a = ""
    
    for i,l in enumerate(s):
        if l != p[i]:
            a += l
            
    return a

def mem_remove(x, l):
    return l.replace(x,"",1)

def color_with_availables(p, s, a):
    c = ""
    
    for i,l in enumerate(p):
        if l == s[i]:
            c += 'g'
        elif l in a:
            c += 'o'
            a = mem_remove(l, a)
        else:
            c += 'b'
            
    return c
            
def color(p, s):
    return color_with_availables(p, s, availables(p, s))

def check_colors(p, o, c):
    for i,v in enumerate(c):
        if (v == 'g' and p[i] != o[i]) or (v != 'g' and p[i] == o[i]):
            return False
    return True

def must_be_present(o, c):
    a = ""
    
    for i,v in enumerate(c):
        if v == 'o':
            a += o[i]
            
    return a

def forbidden(o, c):
    f = ""
    
    for i,v in enumerate(c):
        if v == 'b' and o[i] not in f:
            f += o[i]
            
    return f

def remove_green(p, c):
    r = ""
    
    for i,v in enumerate(c):
        if v != 'g':
            r += p[i]
    
    return r

def check_presence_and_forbidden(p, a, f):
    for l in a:
        if (l in f and a.count(l) != p.count(l)) or (l not in f and a.count(l) > p.count(l)):
            return False
        
    for l in f:
        if l not in a and l in p:
            return False
    
    return True

def check_one(p, o, c):
    if not check_colors(p, o, c):
        return False
    
    a = must_be_present(o, c)
    f = forbidden(o, c)
    r = remove_green(p, c)
    
    return check_presence_and_forbidden(r, a, f)
```

```python
possible_colors = {}
for w in words:
    if w not in possible_colors:
        possible_colors[w] = []
        
    for s in words:
        if s != w:
            c = color(w, s)
            if c not in possible_colors[w]:
                possible_colors[w].append(c)
    
#possible_colors
```

```python
f = open("possible_colors.txt", "w")

f.write("Word\tPossible colors\n")
for w in possible_colors:
    f.write(w)
    f.write("\t[")
    f.write(possible_colors[w][0])
    for i in range(1, len(possible_colors[w])):
        f.write(", ")
        f.write(possible_colors[w][i])
    f.write("]\n")
```

```python
probability = {}

for i,w in enumerate(possible_colors):
    if w not in probability:
        probability[w] = {}
    
    for c in possible_colors[w]:
        if c not in probability[w]:
            probability[w][c] = 0
            
        for cmp in words:
            if check_one(cmp, w, c):
                probability[w][c] += 1
                
        probability[w][c] = probability[w][c] / len(words)
    if (i + 1) % 10 == 0:
        print("Done : ", i + 1, " / ", len(words))

#print(probability)
```

```python
f = open("probability.txt", "w")

f.write("Word\tColor\tMatches\tProbability\n")
for w in probability:
    for c in probability[w]: 
        f.write(w)
        f.write("\t")
        f.write(c)
        f.write("\t")
        f.write(str(probability[w][c] * len(words)))
        f.write("/")
        f.write(str(len(words)))
        f.write("\t")
        f.write(str(probability[w][c]))
        f.write("\n")
    f.write("\n")
```

```python
import math

log2 = {}

for w in probability:
    if w not in log2:
        log2[w] = {}
        
    for c in probability[w]:
        if c not in log2[w]:
            if probability[w][c] > 0:
                log2[w][c] = math.log2(1/probability[w][c])
            else:
                log2[w][c] = 0.0

print(log2)
```

```python
f = open("probability.txt", "w")

f.write("Word\tColor\tMatches\tProbability\tLog2\n")
for w in probability:
    for c in probability[w]: 
        f.write(w)
        f.write("\t")
        f.write(c)
        f.write("\t")
        f.write(str(probability[w][c] * len(words)))
        f.write("/")
        f.write(str(len(words)))
        f.write("\t")
        f.write(str(probability[w][c]))
        f.write("\t\t")
        f.write(str(log2[w][c]))
        f.write("\n")
    f.write("\n")
```

```python
information = {}

for w in probability:
    if w not in information:
        information[w] = 0
        
    for c in probability[w]:
        if probability[w][c] > 0:
            information[w] += probability[w][c] * log2[w][c]
    information[w] = round(information[w], 4)
            
print(information)
```

```python
f = open("info.txt", "w")

info = dict(sorted(information.items(), key=lambda item: item[1], reverse=True))

f.write("Word\tInformation\n")
for w in info:
    f.write(w)
    f.write("\t")
    f.write(str(info[w]))
    f.write("\n")
```

```python
f = open("word_info.pl", "w")

for w in info:
    f.write("word([")
    f.write(w[0])
    for i in range(1, len(w)):
        f.write(", ")
        f.write(w[i])
    f.write("]).\n")
```
